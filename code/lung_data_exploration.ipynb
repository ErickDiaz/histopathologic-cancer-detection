# Lung Histopathological images Data Exploration


```python
import numpy as np
import pandas as pd
import seaborn as sns
from PIL import Image
import matplotlib.pyplot as plt
import os
```

```python
DATA_PATH = "../data/lung_colon_image_set/lung_image_sets/"
```

The subfolder lung_image_sets contains three secondary subfolders:
- lung_aca subfolder with 5000 images of lung adenocarcinomas
- lung_scc subfolder with 5000 images of lung squamous cell carcinomas
- lung_n subfolder with 5000 images of benign lung tissues.

# Lung adenocarcinomas

```python
img_path = DATA_PATH+'lung_aca'
training_data = os.listdir(img_path)
```

```python
image_path = img_path + "/" + training_data[0]
```

```python
img = Image.open(image_path)
image = np.array(img)
```

```python
image.shape
```

```python
fix, ax = plt.subplots(nrows=1, ncols=5, figsize=(15,8))

for i, image in enumerate(training_data[:5]):
    image_path = img_path + "/" + image
    ax[i].imshow(Image.open(image_path))
    
```
